```python
import pandas as pd
import numpy as np

df = pd.read_excel("Rotas.xlsx")
```

```python
df["Distancia_Km"] = df["Distancia_Km"].replace({-100.00: 0})
print(df["Distancia_Km"])
```

```python
df['Distancia_Km'] = df['Distancia_Km'].astype(float)                           
print(df["Distancia_Km"])
```

```python
media = df['Distancia_Km'].mean()
print()
```

```python
df["Distancia_Km"] = df["Distancia_Km"].replace({0: media})
print(df["Distancia_Km"])
```

```python
df['Distancia_Km'] = df['Distancia_Km'].round(2)   

display(df)
```

```python
df_dropped = df.drop(3)
print(df_dropped)
```

```python
df_dropped.to_excel("Rotas_trat.xlsx", index=False)
```
