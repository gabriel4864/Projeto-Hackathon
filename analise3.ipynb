```python
import pandas as pd
df = pd.read_excel("Logística.xlsx")

# Criando uma coluna de atraso: tempo real - tempo previsto
df['Atraso'] = df['Tempo_real'] - df['Tempo_previsto']

# Agrupando por Origem, Destino e Filial_responsavel
resultado = (
    df.groupby(['Filial_responsavel'])
    .agg(
        Tempo_previsto_medio=('Tempo_previsto', 'mean'),
        Tempo_real_medio=('Tempo_real', 'mean'),
        Atraso_medio=('Atraso', 'mean')
    )
    .reset_index()
    .sort_values(by='Atraso_medio', ascending=False)
)
df3 = resultado.head(10)

# Exibindo os 10 maiores custos médios
display(df3)
```

```python
df3.to_excel("Filiais_mais_atrasos.xlsx", index=False)
```
